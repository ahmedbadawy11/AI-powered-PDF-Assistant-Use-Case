```python
from langchain_community.document_loaders.text import TextLoader

loader = TextLoader('all_text.txt')
all_text=loader.load()
```

```python
all_text[0]
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    # Set a really small chunk size, just to show.
    chunk_size=400,
    chunk_overlap=20,
    length_function=len,
    # is_separator_regex=False,
)
```

```python
query_1="Enviromental_factors_egypt_arabic"
meta_data={"document":query_1}
```

```python
chunks = text_splitter.create_documents([all_text[0].page_content],metadatas=[meta_data])
```

```python
chunks[0]
```

```python
from langchain.embeddings import SentenceTransformerEmbeddings
```

```python
from langchain.embeddings import SentenceTransformerEmbeddings

model_path = "./paraphrase_multilingual_MiniLM_model"  # Path where you saved the model files
model_name = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
encode_kwargs = {'normalize_embeddings': True}
llm_embedding = SentenceTransformerEmbeddings(model_name=model_name,encode_kwargs=encode_kwargs)

# Now you can use llm_embedding for your tasks
```

```python
from langchain.vectorstores import Chroma
# vector database
save_to_dir = "wiki_chroma_db_3"
docs_ids=list(range(len(chunks)))
docs_ids=[str(d) for d in docs_ids]
vector_db = Chroma.from_documents(
    chunks,
    llm_embedding,
    persist_directory=save_to_dir,
    ids=docs_ids
)
```

```python
# !pip install accelerate
```

```python
from transformers import AutoTokenizer, AutoModel
from transformers import AutoModelForCausalLM, AutoTokenizer
from transformers.utils.quantization_config import BitsAndBytesConfig


MODEL_NAME = "AhmedBadawy11/Qwen2-7B-Instruct-Environmental-RAG"
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True, bnb_4bit_quant_type="nf4", bnb_4bit_compute_dtype=torch.bfloat16
)

Qwen_tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)

Qwen_model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME, quantization_config=quantization_config, device_map="auto"
)
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains.question_answering import load_qa_chain
```

```python
qna_template = "\n".join([
   "أجب عن السؤال التالي باللغة العربية باستخدام السياق المقدم",
    "إذا لم تكن الإجابة واردة في السياق، فقل'لا توجد إجابة متاحة'",
    "### Context:",
    "{context}",
    "",
    "### Question:",
    "{question}",
    "",
    "### Answer:",
])


qna_prompt=PromptTemplate(
    template=qna_template,
    input_variables=["context","question"],
    # verbose=True
)
```

```python
question = "ما هي التحديات البيئية التي تواجه مصر؟"
query_text = "In which year did Tesla's fathre die?"


similar_docs=vector_db.similarity_search(question,k=4)
```

```python
similar_docs[3]
```

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline
from langchain import HuggingFacePipeline
# Create a QA pipeline
qa_pipeline = pipeline("text-generation", model=Qwen_model , tokenizer=Qwen_tokenizer,max_new_tokens=500 )

# Integrate with langchain
hf_pipeline = HuggingFacePipeline(pipeline=qa_pipeline)
```

```python
stuff_chain = load_qa_chain(hf_pipeline, chain_type="stuff", prompt=qna_prompt)
```

```python
answer=stuff_chain(
    {
        "input_documents":similar_docs,
        "question": question
    },
     return_only_outputs=True,

)
print(answer["output_text"])
```
